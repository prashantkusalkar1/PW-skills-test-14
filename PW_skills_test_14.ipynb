Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

1. Point Anomaly

A point anomaly occurs when a single data point is very different from the rest of the dataset.

2. Contextual Anomaly

A contextual anomaly occurs when a data point is abnormal in a specific context (such as time, season, or location) but may be normal in another context.

3. Collective Anomaly

A collective anomaly occurs when a group of data points together behaves abnormally, even though individual points may appear normal.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

1. Isolation Forest

Approach:
Isolation Forest is a tree-based algorithm that isolates anomalies instead of profiling normal data.

It randomly selects a feature and split value to divide the data.

Anomalies are easier to isolate because they are rare and different, so they require fewer splits in the tree.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:
DBSCAN is a density-based clustering algorithm that groups points in high-density regions and marks points in low-density areas as outliers (noise).

3. Local Outlier Factor (LOF)

Approach:
LOF is a density-based anomaly detection method that compares the local density of a point to the density of its neighbors.

If a point has much lower density than its neighbors, it is considered an outlier.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

1. AR (AutoRegressive) Model

Structure:
The AR model predicts future values based on previous values of the time series.

2. MA (Moving Average) Model

Structure:
The MA model predicts values based on past forecast errors (residuals).

3. ARIMA (AutoRegressive Integrated Moving Average)

Structure:
ARIMA combines:

AR (p) → autoregressive terms

I (d) → differencing to make series stationary

MA (q) → moving average terms

4. SARIMA (Seasonal ARIMA)

Structure:
SARIMA extends ARIMA by adding seasonal components

(p,d,q) = non-seasonal components

(P,D,Q) = seasonal components

s = seasonal period (e.g., 12 for monthly data)

5. SARIMAX (Seasonal ARIMA with Exogenous Variables)

Structure:
SARIMAX extends SARIMA by including external variables (exogenous features).

Example external variables:

Weather

Promotions

Economic indicators

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components
(Include your Python code and output in the code box below.)

````python
```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url)

# Convert Month column to datetime and set it as index
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Plot the original time series
plt.figure(figsize=(10,4))
plt.plot(data['Passengers'])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

**Sample Output**

```
1. A line plot showing the AirPassengers dataset from 1949 to 1960.
2. A decomposition chart with four components:
   - Observed (original data)
   - Trend (long-term increase in passengers)
   - Seasonality (yearly repeating pattern)
   - Residual (random fluctuations)
```
````

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

````python
```python id="iso_forest_taxi"
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load a sample NYC Taxi dataset (or any numerical taxi fare dataset)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/nyc-taxi.csv"
data = pd.read_csv(url)

# Select numerical features for anomaly detection
X = data[['fare_amount', 'passenger_count']].dropna()

# Train Isolation Forest model
iso = IsolationForest(contamination=0.05, random_state=42)
labels = iso.fit_predict(X)

# Label anomalies (-1) and normal points (1)
X['anomaly'] = labels

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(X[X['anomaly'] == 1]['fare_amount'],
            X[X['anomaly'] == 1]['passenger_count'],
            c='blue', label='Normal')

plt.scatter(X[X['anomaly'] == -1]['fare_amount'],
            X[X['anomaly'] == -1]['passenger_count'],
            c='red', label='Anomaly')

plt.xlabel("Fare Amount")
plt.ylabel("Passenger Count")
plt.title("Isolation Forest Anomaly Detection (NYC Taxi Data)")
plt.legend()
plt.show()
```

**Sample Output**

```id="iso_output"
A 2D scatter plot where:
- Blue points represent normal taxi trips
- Red points represent detected anomalies (unusual fares or passenger counts)
```
````

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

````python
```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url)

# Convert Month column to datetime and set as index
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Train SARIMA model
# (p,d,q)(P,D,Q,s) = (1,1,1)(1,1,1,12) for monthly seasonality
model = SARIMAX(data['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_values = forecast.predicted_mean

# Plot original data and forecast
plt.figure(figsize=(10,5))
plt.plot(data['Passengers'], label="Observed")
plt.plot(forecast_values, label="Forecast (Next 12 Months)", color='red')

plt.title("SARIMA Forecast for Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()

plt.show()
```

**Sample Output**

```
A time series plot where:
- The blue line shows the original airline passenger data (1949–1960).
- The red line shows the SARIMA forecast for the next 12 months.
- The forecast continues the upward trend and seasonal pattern observed
```
````
